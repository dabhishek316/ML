```python
# Unsupervised Learning demo: K-Means clustering on 2D synthetic data
# We'll generate 3 natural groups and let KMeans discover them (no labels given).

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1) Create unlabeled data with 3 blob-like groups
X, _ = make_blobs(n_samples=240, centers=3, cluster_std=0.80, random_state=7)

# 2) Run K-Means to find 3 clusters
kmeans = KMeans(n_clusters=3, n_init="auto", random_state=7)
labels = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

# 3) Plot the points colored by their discovered cluster (no explicit colors set)
plt.figure(figsize=(7,5))
plt.scatter(X[:,0], X[:,1], c=labels, alpha=0.7)
plt.scatter(centers[:,0], centers[:,1], s=200, marker="X")  # cluster centers
plt.title("Unsupervised Learning: K-Means Clustering (k=3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("Cluster centers (approx):")
print(centers)
```
